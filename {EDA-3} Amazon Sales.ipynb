**3.  What is the distribution of discounted prices vs. actual prices?**

- Their discounted prices are still high, but the discount isn’t always proportional.

**4. How does the average discount percentage vary across categories?**

```python
df.columns
```

```python
df['discount_percentage'].dtypes
```

```python
avg_discount = df.groupby('category')['discount_percentage'].mean()
avg_discount
```

```python
avg_discount = avg_discount.sort_values(ascending=False)
avg_discount
# This helps you easily see which categories offer the most discount on average.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.hist(avg_discount, bins=20, edgecolor='black', color='skyblue')
plt.xlabel('Average Discount Percentage')
plt.ylabel('Number of Categories')
plt.title('Distribution of Avg. Discount % Across Categories')
plt.grid(True)
plt.show()
```

Insights - 
- Most categories offer moderate discounts (30%–60%)
    - The peak (mode) is around 40%–45%.

- This means most product categories have average discounts in that middle range — a common pricing strategy.

- A few categories have very low or very high average discounts
- Some categories offer almost no discounts (0%–10%) — likely premium or stable-price items.
- A few offer extremely high discounts (70%–90%), possibly seasonal or clearance products.

**6. What are the most popular product keywords?**

```python
#Combining all product names into one string
all_names = ' '.join(df['product_name'].astype(str))
```

```python
#Tokenize into individual words
words = all_names.lower().split()
# words
```

```python
from collections import Counter

#Counting  every word frequencies
word_freq = Counter(words)
# word_freq
```

```python
top_keywords = word_freq.most_common(10)
top_keywords
```

```python
#Convert into DataFrame
top_keywords_df = pd.DataFrame(top_keywords, columns=['Keyword', 'Count'])
top_keywords_df
```

```python
top_keywords_df.set_index('Keyword').plot(kind='barh', figsize=(10,6), legend=False, color='orange')
plt.title('Top 10 Most Common Product Keywords')
plt.xlabel('Frequency')
plt.tight_layout()
plt.show()
```

**8. What is the correlation between discounted_price and rating?**

```python
df[['discounted_price', 'rating']].dtypes
```

```python
df_corr = df[['discounted_price', 'rating']].dropna()
df_corr
```

```python
correlation = df_corr['discounted_price'].corr(df_corr['rating'])
print("Correlation between discounted_price and rating:", correlation)
```

- There's a very weak positive correlation between discounted price and rating.
That means:
    - As discounted price increases, ratings slightly increase, but the relationship is not strong or significant.

    - So, product price has almost no real impact on how users rate it.

```python
plt.scatter(df_corr['discounted_price'], df_corr['rating'], alpha=0.5)
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.title('Discounted Price vs Rating')
plt.grid(True)
plt.show()
```

Insights: Discounted Price vs Rating
- No strong relationship visible
    - Ratings mostly fall between 3.5 and 5, regardless of the discounted price.
    - Even low-priced and high-priced products tend to receive similar rating ranges.

- Most rated products are low to mid-price
    - There’s a heavy concentration of products under ₹10,000 with lots of ratings.
    - Higher-priced products (₹30,000+) are fewer and don’t show any special rating pattern.
