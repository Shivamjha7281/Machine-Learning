# Theoretical

**1. What is a Decision Tree, and how does it work?**

- Imagine we are playing a game where we ask Yes/No questions to reach an answer.
- A Decision Tree is just like that — it's a tree-shaped diagram that helps us make a decision based on questions.
- It's called a "tree" because it starts from the top (called the root) and branches out just like a real tree.
- At the end of each branch, we get a final decision (like which fruit to eat, or what movie to watch).
- Example -
    - Let’s say we are trying to decide what to do on Sunday.
    - We start at the top question: “Is it raining?”
        - If Yes, your decision: Stay home.
        - If No, ask the next question: “Do you have homework?”
            - If Yes, finish homework.
            - If No, go play outside.
- A Decision Tree is like a flowchart of Yes/No questions.
- It helps you (or a computer) make smart decisions step by step.
- It starts with a question and ends with an answer.
- Used in many real-life situations: like choosing clothes, food, or even friends to call.

**2.  What are impurity measures in Decision Trees?**

- Imagine we have a jar of candies.
- If all candies are of one color (say, all red), it’s pure — nice and clean.
- But if the jar has red, green, and yellow candies all mixed, it’s impure — mixed up.
- In Decision Trees, we want to split the data in such a way that each group (or jar) is as pure as possible — meaning it contains data of the same type.
- Impurity measures are math tools that help the computer know: “How mixed is this group?”.
- They help the tree decide where to split the data to make the best decision.
- Main Types of Impurity Measures
    - Gini Index - Measures how mixed a group is. Lower Gini = more pure.
    - Entropy - Measures surprise or disorder. Lower Entropy = less mixed.
    - Misclassification Error - Just checks how many are wrongly grouped.
- Impurity means “how mixed” a group is.
- In decision trees, we want to create groups that are pure.
- Impurity measures help decide where to split the data.
- Lower impurity = better split = smarter decision tree.

**4.  What is the mathematical formula for Entropy?**

- Entropy is a key concept in Decision Trees (including in CatBoost, when it uses trees for splitting), and it's used to measure impurity or disorder in a dataset — basically, how mixed the data is.

- Mathematical Formula for Entropy:

- For a binary classification (like Yes/No), the entropy H of a dataset is calculated as:
    - H(p) = -p1log2(p1) - p2log2(p2)
    - Where:
        - p1 is the probability of class 1 (e.g., "Yes")
        - p2 is the probability of class 2 (e.g., "No")
        - log2 is log base 2
- If all the data is from one class (pure), entropy = 0
- If the data is evenly split (maximum mix), entropy = 1 (for binary case)

**5. What is Information Gain, and how is it used in Decision Trees?**

- Entropy = confusion in the data (like mixed Yes/No answers).
- Information Gain = how much confusion you remove when you split the data using a feature.
- Formula:
    - Information Gain=Entropy (before split)−Entropy (after split)
- Decision Trees choose the feature with the highest Information Gain to split the data at each step.
- Goal: Keep splitting the data using features that make it more pure (less mixed) — until the predictions become clear.

**6.  What is the difference between Gini Impurity and Entropy?**

- Entropy measures the amount of information or "confusion" in the data.
- Gini Impurity measures how often a randomly chosen element would be incorrectly classified.
- Entropy = −∑pilog2pi
- Gini = 1 - ∑pi**2
- Gini and Entropy both range from 0 (pure) to 1 (impure) (in binary classification).
- Gini is faster to compute (no logarithms), so it’s used by default in decision trees like CART.
- Entropy is based on information theory (focuses on unpredictability).
- Gini is based on probability of misclassification.
- Both usually give similar results in practice and often choose the same splits.
- Gini is used in CART algorithm (Classification and Regression Trees).
- Entropy is used in ID3 and C4.5 algorithms.

**8.  What is Pre-Pruning in Decision Trees?**

- Pre-Pruning is a technique used to stop the growth of a decision tree early, before it becomes too complex.
- It applies conditions during tree building, such as:
    - Maximum depth
    - Minimum number of samples to split a node
    - Minimum information gain required for a split
- The goal is to prevent overfitting by limiting the tree size before it fully grows.
- Pre-pruning makes the model simpler, faster, and more generalizable.
- Common parameters used for pre-pruning include:
    - max_depth, min_samples_split, min_samples_leaf (in libraries like Scikit-learn)  

**9. What is Post-Pruning in Decision Trees?**

- Post-pruning means growing the full decision tree first, and then removing the parts of the tree that do not provide significant predictive power.
- After the full tree is built, pruning involves cutting back some branches, especially those that:
    - Classify few data points
    - Are based on noisy data
    - Don’t improve accuracy on validation data
- Suppose that a student writing an exam answer with too many unnecessary details. A teacher might prune the extra content to make the answer clearer and more focused — just like pruning a tree to keep only the most important branches. 

**10. What is the difference between Pre-Pruning and Post-Pruning?**

Pre-Pruning :- 
- Pre-Pruning Applied During tree building (before full tree grows).
- Pre-Pruning Stops splitting a node early if it's not useful.
- Pre-Pruning Decision Based On Criteria like max depth, min samples, etc.


Post-Pruning :- 
- Post-Pruning Applied After full tree is built.
- Post-Pruning Grows the full tree first, then removes weak parts
- Post-Pruning Validation set accuracy or pruning cost function

**11.  What is a Decision Tree Regressor?**

- A Decision Tree Regressor is a type of machine learning model used for predicting continuous (numeric) values, not categories or classes.
- A Decision Tree Regressor works by Splitting the dataset into smaller and smaller parts (like branches on a tree).
- A Decision Tree Regressor works Based on features (columns) that help best predict the target (numeric value).
- A Decision Tree Regressor works by Until the data in each final group (leaf) is pure enough (i.e., has low variance).

**12. What are the advantages and disadvantages of Decision Trees?**

Advantages :- 
- Easy to understand and visualize — Looks like a flowchart, easy to explain.
- No need for feature scaling — No normalization or standardization needed.
- Handles both numerical and categorical data.
- Requires little data preprocessing — No need for encoding or scaling.
- Captures non-linear relationships — Works well even when data isn’t linearly separable.
- Useful for both classification and regression.
- Can be used for feature selection — Identifies important features during splits.

Disadvantages :-
- Prone to overfitting — Especially if the tree is too deep or unpruned.
- Unstable — Small changes in data can result in a completely different tree.
- Less accurate alone — Usually outperformed by ensemble methods like Random Forest, XGBoost.
- Biased toward dominant classes — Struggles with imbalanced datasets.

**13. How does a Decision Tree handle missing values?**

- Decision Tree Skip the missing value feature while splitting — It may choose another feature that has complete data.

- Use surrogate splits — If the main splitting feature is missing, the tree can use a backup feature that behaves similarly.

- Assign missing values to the most frequent category (for categorical data) or mean/median (for numerical data).

- Send missing values down both branches — Weighted by how many training samples went each way (used in some algorithms).

- Some implementations (like sklearn) require you to fill missing data first — e.g., using imputation (SimpleImputer) before training.

**14. How does a Decision Tree handle categorical features?**

- Splits Based on Categories - It asks questions like: "Is Brand = Honda?" or "Is Fuel Type = Petrol?"

- Groups Categories for Best Split - If a feature has multiple categories, the tree finds the best way to group them (e.g., [Honda, Yamaha] vs. [Royal Enfield, Bajaj]).

- Uses Gini or Entropy (for classification) - It calculates which split gives the purest groups based on the target label.

- No Need to Convert to Numbers - Unlike linear models, trees can handle text labels directly (some implementations like sklearn may need encoding, though).

- Handles High Cardinality with Care - If there are too many categories, the tree might overfit, so it may limit or combine them for better performance.

**15. What are some real-world applications of Decision Trees?**

- Medical Diagnosis - Used to predict diseases based on symptoms, age, and test results (e.g., cancer detection, diabetes risk).

-  Loan Approval in Banks - Helps decide whether to approve or reject a loan based on income, credit score, and repayment history.

-  Customer Churn Prediction - Used by telecom and subscription companies to find which customers are likely to leave.

-  Used Car/Bike Price Prediction - Predicts the resale value based on model, age, brand, and kilometers driven.

-   Product Recommendation in E-commerce - Suggests products to users by analyzing browsing and purchase behavior.

**23. Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its accuracy with unscaled data.**

```python
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


clf_unscaled = DecisionTreeClassifier(random_state=42)
clf_unscaled.fit(X_train, y_train)
y_pred_unscaled = clf_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


clf_scaled = DecisionTreeClassifier(random_state=42)
clf_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)


print("Accuracy on unscaled data: {:.2f}%".format(accuracy_unscaled * 100))
print("Accuracy on scaled data: {:.2f}%".format(accuracy_scaled * 100))
```

**24.  Write a Python program to train a Decision Tree Classifier using One-vs-Rest (OvR) strategy for multiclass classification.**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score


iris = load_iris()
X = iris.data
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


ovr_model = OneVsRestClassifier(DecisionTreeClassifier(random_state=42))
ovr_model.fit(X_train, y_train)


y_pred = ovr_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of Decision Tree with One-vs-Rest strategy: {:.2f}%".format(accuracy * 100))
```

**27.  Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(random_state=0)
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

clfs = []
for alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

train_scores = [accuracy_score(y_train, clf.predict(X_train)) for clf in clfs]
test_scores = [accuracy_score(y_test, clf.predict(X_test)) for clf in clfs]
```

```python
# Ploting results
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_scores, marker='o', label='Train Accuracy', drawstyle="steps-post")
plt.plot(ccp_alphas, test_scores, marker='o', label='Test Accuracy', drawstyle="steps-post")
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Effect of CCP Alpha on Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

**28.  Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision, Recall, and F1-Score.**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

report = classification_report(y_test, y_pred, target_names=iris.target_names)
print("Classification Report:\n")
print(report)
```

**29.  Write a Python program to train a Decision Tree Classifier and visualize the confusion matrix using seaborn.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

iris = load_iris()
X, y = iris.data, iris.target
class_names = iris.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```

**30.  Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split.**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

dt = DecisionTreeClassifier(random_state=42)

param_grid = {
    'max_depth': [2, 3, 4, 5, 6, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
