**1. What is Logistic Regression, and how does it differ from Linear Regression.**

Ans - 

- Logistic Regression is a method used to predict something that has only two possible answers — like YES or NO, WIN or LOSE, PASS or FAIL.

- It gives us the probability of something happening, and the final result is usually 0 or 1 (like NO or YES).

-  Linear Regression → Predicts numbers (like 75 marks, 160 cm height).

- Logistic Regression → Predicts yes/no type answers (like Pass/Fail, Buy/Not Buy).

- Imagine we're at a fruit shop, and we're trying to figure out something about apples.

- Linear Regression:
    - we want to predict the weight of an apple just by looking at its size.
    - Input: Size of the apple (small, medium, large)
    - Output: Weight in grams (like 100g, 150g, 200g)
    - This is Linear Regression, because the output is a number.
- Logistic Regression:
    - Now, we want to know that "Will this apple be sweet or not?" based on its color (red or green).
    - Input: Color of the apple
    - Output: Sweet = 1 (Yes), Not Sweet = 0 (No)
    -  This is Logistic Regression, because the answer is Yes/No (1 or 0).

**3. Why do we use the Sigmoid function in Logistic Regression.**

Ans - 

- We use the sigmoid function because it squeezes any number into a range between 0 and 1, like turning a big number into a probability (chance).

- Example: Movie Ticket or Not?
    - Suppose that  a kid wants to know: "Will mom say YES or NO to letting me go to a movie?"
    - He looks at one thing:
        - Did I clean my room today? (If yes → more chance of “Yes” from mom)
        - Now we want to predict:
    - Will mom say YES (1) or NO (0)?
    - We can make a small score (z) based on:
        - +2 points for cleaning room
        - -3 points for playing video games too long
    - That score (z) could be any number: -4, 0, 3, 10, etc. But we can’t give that number directly as a prediction, because… Mom doesn’t say “score = 4”, she says “Yes” or “No”!

- That’s where the Sigmoid function comes in! It turns this score (z) into a probability between 0 and 1.
- Example:
    - If z = 0 - probability = 0.5 (50-50 chance)
    - If z = 2 - probability = 0.88 (88% chance of “Yes”)
    - If z = -3 - probability = 0.04 (only 4% chance of “Yes”)
- So now we can say that "There is an 88% chance that mom will say YES to the movie."

We use the Sigmoid function in Logistic Regression because it takes any number and turns it into a probability — like saying how likely something is to happen (Yes or No).

**6. Explain the difference between Lasso, Ridge, and Elastic Net regression.**

Ans - 

- All three are regularization techniques used to reduce overfitting in regression models.

- Ridge Regression (L2) adds a penalty on the square of the coefficients — it shrinks them but does not make them zero.

- Lasso Regression (L1) adds a penalty on the absolute value of the coefficients — it can shrink some to exactly zero, removing unimportant features.

- Elastic Net is a combination of Ridge + Lasso — it can shrink and also eliminate features.

- Ridge is good when all features are somewhat useful.

- Lasso is useful when you want feature selection (automatically choosing only the important ones).

- Elastic Net works well when you have many features and some are correlated (related to each other).

**7. When should we use Elastic Net instead of Lasso or Ridge?**

Ans - 

- We have many features (inputs) — especially more than the number of data points (rows in your dataset).

- Some features are highly correlated (i.e., they give similar information — like "height in cm" and "height in inches").

- Lasso alone is unstable — it may randomly pick one feature and ignore the other, even if both are useful. Elastic Net handles this better by combining L1 + L2.

- If we want both benefits then :
    - Feature selection like Lasso (remove useless features)
    - Stable coefficients like Ridge (for correlated features)

- Example - Imagine we're predicting house prices using 100 features, like:
    - Number of rooms
    - Area in sq ft
    - Distance to school
    - Distance to supermarket
    - Distance to park (these 3 may be correlated)
-  Elastic Net is best here — it:
    - Removes unhelpful features,
    - Keeps the model stable if some features are similar.

**9. What are the key assumptions of Logistic Regression.**

Ans - 

- Binary Output (for Binary Logistic Regression) The outcome (target variable) should be binary — like 0 or 1, Yes or No, Pass or Fail.

- Independence of Observations Each observation (row of data) should be independent of the others.
    - Example: One person's result shouldn't affect another's.

- No (or little) Multicollinearity
    - The independent variables (inputs) should not be highly correlated with each other.
    - Example: Don’t include both "height in cm" and "height in inches" — they carry the same info.

- Linearity of Logit
    - Logistic Regression assumes a linear relationship between the log-odds of the output and the input variables. (Not the output itself — but the log-odds.)
 
- Large Sample Size
    - Logistic Regression performs better with enough data. It can struggle with very small datasets.
 
- No extreme outliers
    - Outliers can affect model performance, especially in smaller datasets. Handling them improves reliability.
 
- Predictors are measured correctly Input features (X) should be accurate and ideally numeric. Categorical features should be properly encoded (like One-Hot Encoding).

**11. What are Classification Evaluation Metrics?**

Ans - 

- They are tools or formulas that help you check - "How good is my model at predicting the correct class (like Yes/No, 1/0, Pass/Fail)?"

- Most Common Classification Metrics:
    - Accuracy – Tells how many total predictions were correct (good for balanced data).
    - Precision – Out of all predicted "Yes" (positive), how many were actually correct (important when false positives are costly).
    - Recall – Out of all actual "Yes" cases, how many the model correctly found (important when missing real cases is bad).
    - F1 Score – A balanced score combining both precision and recall (used when data is imbalanced).
    - Confusion Matrix & ROC-AUC – Tools to see detailed prediction performance and how well the model separates the classes.

**12. How does class imbalance affect Logistic Regression?**

class imbalance :
- It happens when the number of examples in one class (like "Yes") is much smaller or larger than the other class (like "No").
- Example:
    - 950 customers didn’t buy (class = 0)
    - 50 customers bought (class = 1)
    - This is a 95:5 imbalance.

How Class Imbalance Affects Logistic Regression:
- Biased Predictions - Logistic Regression tends to favor the majority class (predicting 0 most of the time). It might predict "No" for everyone — and still get 95% accuracy.

- Misleading Accuracy - High accuracy doesn’t mean good performance. The model can miss almost all the minority cases (like buyers, fraud, or diseases).

- Poor Recall for Minority Class - Logistic Regression may struggle to detect the rare class, resulting in low recall (missing real positive cases).

**15. How is Logistic Regression extended for multiclass classification?**

Ans - 

- Logistic Regression is naturally built for binary classification (Yes/No, 0/1), but it can be extended for multiclass problems (like predicting A, B, or C).


- There are two main strategies used:
    - One-vs-Rest (OvR) (also called One-vs-All)
        - The model creates one binary classifier for each class.
        - For example, if we have 3 classes (A, B, C):
            - Model 1: A vs (B and C)
            - Model 2: B vs (A and C)
            - Model 3: C vs (A and B)
            - The model gives a probability for each class, and the one with the highest score wins.
    - Multinomial (Softmax Regression)
        - A single model is trained to predict all classes at once.
        - Uses the softmax function to calculate the probability of each class.
        - More accurate when classes are mutually exclusive.
        - It Works well with ‘lbfgs’ or ‘saga’ solvers.

- Example:
    - we predicting which fruit a person will pick:
        - Apple
        - banana
        - grapes
    - OvR: Train 3 binary models: Apple vs Not Apple, Banana vs Not Banana, Grapes vs Not Grapes
    - Multinomial: Train one model that directly gives probabilities like:
        - Apple: 0.2, Banana: 0.3, Grapes: 0.5 - wins   

**16. What are the advantages and disadvantages of Logistic Regression?**

Ans - 

Advantages of Logistic Regression:
- Simple and Easy to Implement - Logistic Regression is easy to understand and implement, especially for binary classification problems.
- Interpretable - Coefficients represent the effect of each feature on the outcome (odds). This makes it very useful for insights and reporting.
- Efficient on Linearly Separable Data - Performs well when the classes are linearly separable and the relationship between features and the target is linear (in log-odds).
- Fast Training - Logistic Regression is computationally less intensive and trains quickly even on large datasets.
- Works Well with Less Noise - Performs reliably if the dataset has minimal noise or outliers.

Disadvantages of Logistic Regression:
- Assumes Linear Relationship - Assumes a linear relationship between independent variables and the log-odds of the dependent variable, which may not always hold.
- Not Suitable for Complex Relationships - Can't capture non-linear relationships unless features are transformed or interaction terms are added.
- Sensitive to Multicollinearity - If features are highly correlated, the model's coefficients become unstable and hard to interpret.

**18. What is the difference between Softmax Regression and Logistic Regression?**

Ans - 

- Type of Problem
    - Logistic Regression: Used for binary classification (2 classes).
    - Softmax Regression: Used for multiclass classification (3 or more classes).
- Output
    - Logistic: Predicts one probability (e.g., probability of class 1).
    - Softmax: Predicts a probability for each class.
- Activation Function
    - Logistic: Uses the sigmoid function.
    - Softmax: Uses the softmax function.
- Prediction Decision
    - Logistic: Class with probability > 0.5 is chosen.
    - Softmax: Class with the highest probability is chosen.
- Formula Type
    - Logistic: Calculates log-odds for one class.
    - Softmax: Calculates exponentiated scores for all classes.
- Common Use Cases
    - Logistic: Spam detection, churn prediction (Yes/No problems).
    - Softmax: Digit recognition, species classification (multi-class problems).

**19.  How do we choose between One-vs-Rest (OvR) and Softmax for multiclass classification?**

Ans - 

- OvR: Makes one model for each class (e.g., Cat vs Not Cat, Dog vs Not Dog).

- Softmax: Makes one single model that chooses the best class directly.

- Number of Models -
    - OvR: One model per class (more models).
    - Softmax: Only one model for all classes.
- Speed
    - OvR: Slower (because it trains many models).
    - Softmax: Faster and more efficient.
- Prediction Style
    - OvR: Each model says “yes or no” for its class. Pick the one with highest score.
    - Softmax: Gives probabilities for all classes in one go.
- Accuracy
    - OvR: Might be less accurate if classes are similar.
    - Softmax: Usually more accurate as it compares all classes together.
- When Classes Are Related
    - OvR: Doesn't know about other classes while training.
    - Softmax: Considers all classes at once, so it's smarter.
- Use When?
    - OvR: Good for few classes or quick models.
    - Softmax: Better for many classes or better performance.
- In Practice
    - In scikit-learn, you can try both:
        - OvR: multi_class='ovr'
        - Softmax: multi_class='multinomial', solver='lbfgs'

**20. How do we interpret coefficients in Logistic Regression?**

Ans - 

What Do Coefficients Represent?
- In logistic regression, each coefficient tells you how a feature affects the log-odds of the outcome.
If the coefficient is:
    - Positive: Increases the chance of the event happening.
    - Negative: Decreases the chance of the event happening.
- But instead of "log-odds", we often convert coefficients into odds ratios for easy interpretation.

Example - Predicting Loan Approval

Suppose that we build a logistic regression model to predict whether a loan will be approved (Yes = 1 / No = 0) based on:

Assume we get the following model:
- log-odds(approval) = -3 + 0.02 * income + 0.01 * credit_score - 0.05 * age 

Interpreting the Coefficients
- Intercept = -3
    - When all features are 0, the base log-odds of approval is -3 (not meaningful by itself, but used in the full formula).
- Income Coefficient = 0.02
    - For every ₹1,000 increase in income, the odds of loan approval increase by about 2%.
(Because: exp(0.02) ≈ 1.02 → 2% increase)
- Credit Score Coefficient = 0.01
    - For every 1-point increase in credit score, the odds of approval increase by ~1%.
(exp(0.01) ≈ 1.01)
- Age Coefficient = -0.05
    - For every 1-year increase in age, the odds of approval decrease by ~5%.
(exp(-0.05) ≈ 0.95)

**7. Write a Python program to evaluate Logistic Regression using Stratified K-Fold Cross-Validation. Print the average accuracy.**

```python
from sklearn.model_selection import StratifiedKFold
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
accuracies = []
```

```python
for train_index, test_index in skf.split(X, y):
    # Split data
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train model
    model = LogisticRegression(solver='liblinear')
    model.fit(X_train_scaled, y_train)

    # Predict and evaluate
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

```python
average_accuracy = np.mean(accuracies)
print(f"Average Accuracy with Stratified K-Fold CV: {average_accuracy * 100:.2f}%")
```

**24.  Write a Python program to train Logistic Regression and find the optimal C (regularization strength) using cross-validation.**

```python
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LogisticRegression(solver='liblinear', max_iter=500)

param_grid = {'C': [0.01, 0.1, 0.5, 1, 5, 10]}

grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Best C value:", grid.best_params_['C'])
print("Test Accuracy:", round(accuracy, 4))
```
